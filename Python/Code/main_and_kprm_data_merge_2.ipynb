```python
import pandas as pd
import gdown
```

### Loading data

```python
# Additional KPRM data - candidates.
file_id = "1b7nKk6yu8Owz90QfCRwMRRWXAP2zMhIP"
url = f"https://drive.google.com/uc?id={file_id}"
file_extension = "downloaded_file.csv"
gdown.download(url, file_extension, quiet=False)
kprm_kandydaci = pd.read_csv(file_extension, sep=",")
kprm_kandydaci.head()
```

```python
# Additional KPRM data - employment.
file_id = "1aUf3-XXXoTzW3Ff33mgDHbpp-1elKqSA"
url = f"https://drive.google.com/uc?id={file_id}"
file_extension = "downloaded_file.csv"
gdown.download(url, file_extension, quiet=False)
kprm_zatrudnienie = pd.read_csv(file_extension, sep=",")
kprm_zatrudnienie.head()
```

```python
# Main data.
file_id = "1A71j8RJUzNZes3GkZEAVkj0pEK5OvT71"
url = f"https://drive.google.com/uc?id={file_id}"
file_extension = "downloaded_file.csv"
gdown.download(url, file_extension, quiet=False)
main_df = pd.read_csv(file_extension, sep=",")
main_df.head()
```

### Merging KPRM data part 1 with cleansed main data

```python
kprm_kandydaci.head()
```

```python
# Creating merge fields (institution name in the KPRM data is preprocessed in the same way as below).
main_df["institution_merge_copy"] = main_df["institution"].copy().str.replace(' ', '').str.lower()
main_df["year_announced"] = pd.to_datetime(main_df["date_announced"].copy()).dt.year

# Merging.
initial_first_merge_df = pd.merge(
    main_df, kprm_kandydaci, left_on= ['institution_merge_copy', 'year_announced'], right_on=['podmiot', 'rok'], how='left'
)
initial_first_merge_df.head()
```

```python
# Inspecting the correctness of the merge.
print(
    "Dataframe's total rows: " + str(initial_first_merge_df.shape[0]) + "\n" +
    "Nulls: " + str(initial_first_merge_df["kobiety"].isnull().sum())
)
```

##### Merging

```python
# Some cleaning first.
kprm_zatrudnienie["grupa_stanowisk"] = kprm_zatrudnienie["grupa_stanowisk"].str.lower()

kprm_zatrudnienie["grupa_stanowisk"] = (
    kprm_zatrudnienie["grupa_stanowisk"]
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True) # replaces multiple spaces with a single space
)

# Repeating the process for the other df.
initial_first_merge_df["position_category"] = initial_first_merge_df["position_category"].str.lower()

initial_first_merge_df["position_category"] = (
    initial_first_merge_df["position_category"]
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True) # replaces multiple spaces with a single space
)
```

```python
second_merge_df = pd.merge(
    initial_first_merge_df, kprm_zatrudnienie, left_on= ['institution_merge_copy', 'year_announced', 'position_category'], right_on=['podmiot', 'rok', 'grupa_stanowisk'], how='left'
)
second_merge_df.head()
```

```python
# Inspecting the correctness of the merge.
print(
    "Dataframe's total rows: " + str(second_merge_df.shape[0]) + "\n" +
    "Nulls: " + str(second_merge_df["zatr_etat"].isnull().sum())
)
```

### Cleaning merged data

```python
# Changing names.
second_merge_df["salary_for_position"] = second_merge_df["wynagrodzenie"]
second_merge_df["offer_salary"] = second_merge_df["salary"]
second_merge_df["women_percentage"] = second_merge_df["kobiety"]
second_merge_df["fluctuation"] = second_merge_df["fluktuacja"]
second_merge_df["employment_contract"] = second_merge_df["zatr_etat"]

# Removing duplicated columns.
second_merge_df.drop(columns=[
    "wynagrodzenie", "salary", "institution_merge_copy", "rok_x", "rok_y", "podmiot_x", "podmiot_y", "grupa_stanowisk", "kobiety", "fluktuacja", "zatr_etat"
], inplace=True)
```

```python
second_merge_df.to_csv("merged_data_2.csv", index=False)
```
