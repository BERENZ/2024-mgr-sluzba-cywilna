```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gdown
```

```python
# Loading data from GDrive.
file_id = "1IlyXO5rXkLi28Z-o-WWCjaFa_rLDrKls"
url = f"https://drive.google.com/uc?id={file_id}"
file_extension = "downloaded_file.csv"
gdown.download(url, file_extension, quiet=False)
df = pd.read_csv(file_extension, sep=",")
```

```python
df.head()
```

```python
# Grupowanie i agregowanie danych
df = df[['district', 'institution', 'year_announced', 'result', 'education_level', 'work_time', 'vacancies', 'views', 'promocja', 'kandydaci', 'salary_for_position', 'women_percentage', 'fluctuation', 'employment_contract', 'voivodeship']]

df_grouped = df.groupby(['district', 'voivodeship', 'institution', 'year_announced']).agg('mean').reset_index()
df_grouped.head(100)
```

```python
df_grouped['filled_vacancies_per_all'] = df_grouped['result'] / df_grouped['vacancies']
```

```python
df_grouped.head(100)
```

```python
# Calculate correlation between 'salary_for_position' and 'result'
correlation = df_grouped[['salary_for_position', 'result']].corr()

# Display the correlation matrix
print(correlation)
```
