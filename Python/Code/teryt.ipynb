```python
import pandas as pd
import numpy as np
import gdown
# Google Drive file ID
file_id_kprm = "1ZRScvG_jTStNdNeUGDW077lUVrWKJIY9"
url_kprm = f"https://drive.google.com/uc?id={file_id_kprm}"
file_kprm_extension = "kprm_data.csv"

# Download the file from Google Drive
gdown.download(url_kprm, file_kprm_extension, quiet=False)

# Load the data from the downloaded CSV file
kprm = pd.read_csv(file_kprm_extension, sep=";")
kprm.head()
```

```python
kprm['rok'] = kprm['date_annouced'].str.slice(0, 4).astype(int)
kprm['city_merge'] = kprm['city'].where(~kprm['city'].str.contains('i teren'), 
                                              kprm['city'].str.split(' i teren').str[0])
kprm['city_merge'] = kprm['city_merge'].str.replace(r'[-\s.]|\(.*\)', '', regex=True).str.lower()
kprm = kprm [['rok', 'city_merge','city', 'address', 'work_place1','work_place2', 'job_id']]
kprm
```

```python
# List of Google Drive file IDs
file_ids = [
    "1x8Rk7Cvd29OPQWhgC_mKVCQHmx-qPp1W",  # File 1
    "1XeqLqx-V-voeUQqWCAGH4FGKmyPLrZbf",  # File 2
    "1NTm4j7kyjoCSd1qv-nDB3bBV5ZDe0SjG",  # File 3
    "16JBRzwbTHfCOmLUrPzbo3KlLIsPvSEj8",  # File 4
    "1kT2_tdpQ5usmr10LNd_9BS8C60n_YIMY",  # File 5
    "18IWAGl9O5orChNnSiROFaWQPqU66Z6nS",  # File 6
    "1ZHS77ex5fntFNMjPBDcRp2SJtYWUMoRq"   # File 7
]

# List to store DataFrames
teryt = []

# Download and load CSV files into DataFrames
for file_id in file_ids:
    url = f"https://drive.google.com/uc?id={file_id}"
    output_file = f"{file_id}.csv"  # File name based on ID
    gdown.download(url, output_file, quiet=False)  # Download file
    
    # Read the CSV file into DataFrame
    df = pd.read_csv(output_file, sep=';', dtype={'WOJ': str, 'POW': str})
    teryt.append(df)  # Add DataFrame to list

# Concatenate all DataFrames
terc = pd.concat(teryt, ignore_index=True)
terc['rok'] = terc['STAN_NA'].str.slice(0, 4).astype(int)
terc['city_merge'] = terc['NAZWA'].str.replace(r'[-\s]', '', regex=True).str.lower()
terc = terc[['rok', 'city_merge', 'WOJ', 'POW']]
terc = terc.dropna(subset=['POW'])

# Liczenie wystąpień dla każdej kombinacji 'WOJ' i 'POW'
def most_frequent_group(group):
    most_frequent_woj = group['WOJ'].mode()[0]
    most_frequent_pow = group['POW'].mode()[0]
    return group[(group['WOJ'] == most_frequent_woj) & (group['POW'] == most_frequent_pow)]

# Grouping, filtering, and selecting the most frequent combination of 'WOJ' and 'POW'
terc_filtered = terc.groupby(['rok', 'city_merge'], as_index=False).apply(most_frequent_group)

# Removing duplicates based on 'year' and 'city_merge'
terc_unique = terc_filtered.drop_duplicates(subset=['rok', 'city_merge'])
terc_unique.head()
```

```python
# List of Google Drive file IDs for simc data
file_ids = [
    "1AoGP78dOiabsHnJ63zvjuJk4iwEaGPI1",  # File 1
    "1Bu-qcoPO1bKBMx0hKJqc7PtD97wiLkqW",  # File 2
    "1luUtEJeCUrvMDKt_S9YKnYfsAS0KLVEQ",  # File 3
    "1fp36xfIBThKFlswv2wt5tHrwkiINsM78",  # File 4
    "11dykkAP_tg6RdfEQh0GG7zmA-8RHD-Dh",  # File 5
    "1iJBY8iOUUQoonI6Q04QlGrEiEmN8ZdHj",  # File 6
    "1wYjiwxL-BS-HKkNwe_OytnOUaqbPiLfI"   # File 7
]

# List to store DataFrames
simc = []

# Download and load CSV files into DataFrames
for file_id in file_ids:
    url = f"https://drive.google.com/uc?id={file_id}"
    output_file = f"{file_id}.csv"  # File name based on ID
    gdown.download(url, output_file, quiet=False)  # Download file
    
    # Read the CSV file into DataFrame
    df = pd.read_csv(output_file, sep=';', dtype={'WOJ': str, 'POW': str})
    simc.append(df)  # Add DataFrame to list

# Concatenate all DataFrames
sim = pd.concat(simc, ignore_index=True)
sim['rok'] = sim['STAN_NA'].str.slice(0, 4).astype(int)
sim['city_merge'] = sim['NAZWA'].str.replace(r'[-\s]', '', regex=True).str.lower()
sim = sim[['rok', 'city_merge', 'WOJ', 'POW']]

sim_filtered = sim.groupby(['rok', 'city_merge'], as_index=False).filter(
    lambda group: group['WOJ'].nunique() == 1 and group['POW'].nunique() == 1
)

# Usuwanie duplikatów na podstawie 'rok' i 'city_merge', pozostawiając jedną kombinację
sim_unique = sim_filtered.drop_duplicates(subset=['rok', 'city_merge'])

sim_unique.head()
```

```python
teryt_city = pd.merge(kprm, terc_unique, how='left', on=['city_merge', 'rok'])

# Znalezienie wierszy z brakującymi wartościami WOJ
missing_woj = teryt_city['WOJ'].isnull()

# Połączenie sim_unique TYLKO z wierszami z brakującymi WOJ
teryt_city = pd.merge(teryt_city, sim_unique, how='left', on=['city_merge', 'rok'], suffixes=('', '_sim'))

# Uzupełnienie brakujących wartości WOJ i POW wartościami z sim_unique
teryt_city['WOJ'] = np.where(missing_woj, teryt_city['WOJ_sim'], teryt_city['WOJ'])
teryt_city['POW'] = np.where(missing_woj, teryt_city['POW_sim'], teryt_city['POW'])

# Usunięcie tymczasowych kolumn z sufiksem _sim
cols_to_drop = [col for col in teryt_city.columns if col.endswith('_sim')]
teryt_city = teryt_city.drop(columns=cols_to_drop, errors='ignore')
teryt_city['WOJ'].isna().sum()
```

```python
teryt_city_miss = teryt_city[teryt_city['WOJ'].isnull()].copy()
missing_cities = teryt_city_miss['city_merge'].unique()
```
